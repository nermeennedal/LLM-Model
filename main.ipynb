```python
message_one="Hi , i'm nermeen a computer engineer ! "
massage3="tell me about AI , and how we can learn it "
massage_format=[{"role":"user" ,  "content":message_one}]

massage_second_format=[{"role":"user","content":message_one},
{"role":"user","content":"how are uou  "},
{"role":"user","content":massage3}]


# how the massage is formatted 
```

```python
openai=OpenAI() #we made an instance from OpenAI class

response=openai.chat.completions.create(model="gpt-4.1-nano",messages=massage_second_format,max_tokens=100,n=3)

```
